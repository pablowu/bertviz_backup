```python
model_type = 'xlnet'
model_version = 'xlnet-large-cased'
model = XLNetModel.from_pretrained(model_version)
tokenizer = XLNetTokenizer.from_pretrained(model_version)
text = "At the store she bought apples, oranges"
show(model, model_type, tokenizer, text)
```
